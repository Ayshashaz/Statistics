```python
data = data.drop_duplicates()
```

```python
data['price_per_sqft'] = data['price'] / data['total_sqft']
print(data['price_per_sqft'].describe())
```

```python
print(data['price_per_sqft'].describe())
```

```python
mean = data['price_per_sqft'].mean()
std_dev = data['price_per_sqft'].std()
threshold = 3

outliers_mean_std = data[(data['price_per_sqft'] < mean - threshold * std_dev) |
                       (data['price_per_sqft'] > mean + threshold * std_dev)]


lower_percentile = 0.05
upper_percentile = 0.95
lower_threshold = data['price_per_sqft'].quantile(lower_percentile)
upper_threshold = data['price_per_sqft'].quantile(upper_percentile)

outliers_percentile = data[(data['price_per_sqft'] < lower_threshold) |
                         (data['price_per_sqft'] > upper_threshold)]

Q1 = data['price_per_sqft'].quantile(0.25)
Q3 = data['price_per_sqft'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = data[(data['price_per_sqft'] < lower_bound) |
                  (data['price_per_sqft'] > upper_bound)]

data['z_score'] = (data['price_per_sqft'] - mean) / std_dev
outliers_z_score = data[(data['z_score'] > threshold) | (data['z_score'] < -threshold)]

outliers_summary = {
    'Mean & Std Dev': outliers_mean_std.index.tolist(),
    'Percentile': outliers_percentile.index.tolist(),
    'IQR': outliers_iqr.index.tolist(),
    'Z-Score': outliers_z_score.index.tolist(),
}

outliers_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x=data['price_per_sqft'])
plt.title('Boxplot of Price Per Square Feet Before Outlier Removal')
plt.xlabel('Price Per Square Feet')
plt.show()

data_trimmed_percentile = data[(data['price_per_sqft'] >= lower_threshold) & 
                           (data['price_per_sqft'] <= upper_threshold)]

plt.figure(figsize=(10, 6))
sns.boxplot(x=data_trimmed_percentile['price_per_sqft'])
plt.title('Boxplot of Price Per Square Feet After Outlier Removal (Percentile Method)')
plt.xlabel('Price Per Square Feet')
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis


plt.figure(figsize=(10, 6))
sns.histplot(data['price_per_sqft'].dropna(), kde=True)
plt.title('Histogram of Price Per Square Feet Before Transformation')
plt.xlabel('Price Per Square Feet')
plt.ylabel('Frequency')
plt.show()

skewness_before = skew(data['price_per_sqft'].dropna())
kurtosis_before = kurtosis(data['price_per_sqft'].dropna())

print(f'Skewness Before Transformation: {skewness_before}')
print(f'Kurtosis Before Transformation: {kurtosis_before}')

data['price_per_sqft_log'] = np.log1p(data['price_per_sqft'].dropna())

plt.figure(figsize=(10, 6))
sns.histplot(data['price_per_sqft_log'], kde=True)
plt.title('Histogram of Price Per Square Feet After Log Transformation')
plt.xlabel('Log of Price Per Square Feet')
plt.ylabel('Frequency')
plt.show()

skewness_after = skew(data['price_per_sqft_log'].dropna())
kurtosis_after = kurtosis(data['price_per_sqft_log'].dropna())

print(f'Skewness After Transformation: {skewness_after}')
print(f'Kurtosis After Transformation: {kurtosis_after}'
```
